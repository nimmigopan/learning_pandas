##  how to fetch info for a particular date every year... eg) 6th jan 2004 to2022

```python
pd.date_range(start='2004-01-06',end='2022-01-06',freq=pd.Timedelta(years=1))

# allowed keywords are [weeks, days, hours, minutes, seconds, milliseconds, microseconds, nanoseconds].. years not allowed
```

```python
pd.date_range(start='2004-01-06',end='2022-01-06',freq=pd.Timedelta(days=365))  #this will print every 365th date

# ans is not correct.. bcoz of leapyear, calculation changes
```

```python
# Another object "DateOffset" used instead

pd.date_range(start='2004-01-06',end='2022-01-06',freq=pd.DateOffset(years=1))
```

```python
# to display data corresponding to this dates only,

google[google.index.isin(pd.date_range(start='2004-01-06',end='2022-01-06',freq=pd.DateOffset(years=1)))]
```

## Resampling

Resampling involves changing the frequency of your time series observations.

Two types of resampling are:

* Upsampling: Where you increase the frequency of the samples, such as from minutes to seconds.
* Downsampling: Where you decrease the frequency of the samples, such as from days to months

if ur data is too noisy with many spikes, downsampling will helps to analyse the pattern

```python
google['Close'].plot()
```

```python
google['Close'].resample('W').mean().plot()
```

```python
google['Close'].plot()
google['Close'].resample('W').mean().plot()
```

```python
google['Close'].plot()
google['Close'].resample('M').mean().plot()
```

```python
google['Close'].plot()
google['Close'].resample('Q').mean().plot()
```

```python
#upsampling

google['Close'].resample('H').interpolate(method='linear').plot() # H->hourly.. during upsampling, there will be missing values..fill it with some methods
```
